```python
import numpy as np
from PIL import Image
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from kneed import KneeLocator
from IPython.display import Image as imm
```

```python
img=Image.open('/content/Image/dash.JPG')
```

```python
img
```

```python
# Function to get the best k for the image
def get_k(image,print_plot=False):
    arr=np.array(image)
    arr=arr.reshape(image.size[0]*image.size[1],3)
    wcss=[]
    for i in range(1,11):
        kmeans = KMeans(n_clusters= i, init='k-means++', random_state=0)
        kmeans.fit(arr)
        wcss.append(kmeans.inertia_)
    x=[i for i in range(1,11)]
    kn = KneeLocator(x, wcss, curve='convex', direction='decreasing')
    if print_plot==True:
        plt.xlabel('number of clusters k')
        plt.ylabel('Sum of squared distances')
        plt.plot(x, wcss, 'rx-')
        plt.vlines(kn.knee, plt.ylim()[0], plt.ylim()[1], linestyles='dashed')
        plt.show()
        print('The elbow is formed at :',kn.knee)
    else:
        return kn.knee
```

```python
get_k(img,print_plot=True)
```

```python
def print_dominant_colors(image,custom_k=0):
    arr=np.array(image)
    arr=arr.reshape(image.size[0]*image.size[1],3)
    if custom_k==0:
        kg=get_k(image)
    else:
        kg=custom_k
    kmeans=KMeans(n_clusters=kg, init='k-means++', random_state=0).fit(arr)
    clusters=kmeans.predict(arr)
    
    
    centroids=kmeans.cluster_centers_.astype(int)
    
    k=len(centroids)
    fig, axs = plt.subplots(1, k)
    fig.set_size_inches(3 * k, 3, forward=True)
    for clusternumber in range(k):
      centroid = np.zeros((100,100,3))
      centroid[:,:,:] = centroids[clusternumber]
      axs[clusternumber].imshow(centroid/255)
      axs[clusternumber].text(50,50,str(clusternumber),fontsize=20)
      axs[clusternumber].axis("off")
    plt.show()
```

```python
print_dominant_colors(img)
```

```python
print_dominant_colors(img,custom_k=7)
```

```python
def pie_maker(image,custom_k=0):
    arr=np.array(image)
    arr=arr.reshape(image.size[0]*image.size[1],3)
    if custom_k==0:
        kg=get_k(image)
    else:
        kg=custom_k
    kmeans=KMeans(n_clusters=kg, init='k-means++', random_state=0).fit(arr)
    clusters=kmeans.predict(arr)
    centroids=kmeans.cluster_centers_.astype(int)
    labels=kmeans.labels_
    labels=list(labels)
    percent=[]
    for i in range(len(centroids)):
        j=labels.count(i)
        j=j/(len(labels))
        percent.append(j)
    plt.figure(figsize=(18,8))
    plt.pie(percent,colors=np.array(centroids/255),labels=np.arange(len(centroids)))
    plt.show()
```

```python
pie_maker(img)
```

```python
pie_maker(img,20)
```
